```python
import pandas as pd
```

```python
def load_311(filename=None, rows=1e5, chunk=True):
    bad_columns = [
        'Street Name', 'Cross Street 1', 'Cross Street 2',
        'X Coordinate (State Plane)', 'Y Coordinate (State Plane)',
        'Open Data Channel Type', 'Park Facility Name', 'Park Borough',
        'Vehicle Type', 'Taxi Company Borough', 'Taxi Pick Up Location',
        'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp',
        'Bridge Highway Segment', 'BBL', 'Community Board', 'Landmark',
        'Location', 'Resolution Description', 'Resolution Action Updated Date',
        'Incident Address', 'Intersection Street 1', 'Intersection Street 2',
        'Unique Key', 'Closed Date', 'Agency Name', 'Status', 'Due Date'
    ]

    # read the header to get the column names
    _df = pd.read_csv(filename, nrows=1)

    columns = _df.columns
    good_columns = [col for col in columns if col not in bad_columns]
    print(good_columns)
    
    if chunk:
        chunks = []
        # load the big file in smaller chunks
        for df_chunk in pd.read_csv(filename,
                                    chunksize=rows,
                                    usecols=good_columns):
            df_chunk = df_chunk[df_chunk.Agency.isin(['DEP', 'DOHMH', 'DSNY'])]
            #             print(df_chunk.shape, df_chunk.memory_usage().sum()/1000)
            chunks.append(df_chunk)

        # put the chunks back together
        return pd.concat(chunks, ignore_index=True)

    else:
        return pd.read_csv(filename, nrows=rows, usecols=good_columns)
```
